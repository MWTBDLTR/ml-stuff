# Data Exploration: Understanding Your Dataset

In this notebook, you'll learn how to:
- Load datasets using scikit-learn
- Explore dataset structure and statistics
- Visualize data to understand distributions and relationships
- Identify features and target variables

This is the first step in any machine learning project: understanding your data!


## 1. Import Required Libraries

Let's start by importing the libraries we'll need for data exploration and visualization.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, load_digits, fetch_california_housing

# Configure matplotlib for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

## 2. Load Sample Dataset

Let's load the **Iris Dataset**, a classic dataset used in machine learning. It contains measurements of iris flowers.


```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (measurements)
y = iris.target  # Target (species)
feature_names = iris.feature_names
target_names = iris.target_names

# Create a pandas DataFrame for easier exploration
df = pd.DataFrame(X, columns=feature_names)
df['Species'] = iris.target_names[y]

print("Dataset loaded successfully!")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Feature names: {list(feature_names)}")
print(f"Target classes: {list(target_names)}")
```

## 3. Explore Dataset Structure

Let's examine the dataset more closely by looking at basic statistics and the first few rows.


```python
# Display first few rows
print("First 5 rows of the dataset:")
print(df.head())
print("\n" + "="*60 + "\n")

# Display data types
print("Data types:")
print(df.dtypes)
print("\n" + "="*60 + "\n")

# Display statistical summary
print("Statistical Summary:")
print(df.describe())
```

## 4. Class Distribution

Let's see how many samples of each species we have in our dataset.


```python
# Count samples per species
print("Class distribution:")
print(df['Species'].value_counts())
print()

# Visualize class distribution
plt.figure(figsize=(8, 5))
df['Species'].value_counts().plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Distribution of Iris Species', fontsize=14, fontweight='bold')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Feature Distributions

Let's visualize how each feature is distributed across the dataset.


```python
# Plot histograms for each feature
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Feature Distributions', fontsize=16, fontweight='bold')

for idx, (ax, feature) in enumerate(zip(axes.flat, feature_names)):
    ax.hist(df[feature], bins=20, color='steelblue', alpha=0.7, edgecolor='black')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')
    ax.set_title(f'{feature}')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Feature Relationships

Let's examine relationships between features using scatter plots. We'll focus on two features at a time.


```python
# Create scatter plot: Sepal Length vs Petal Length, colored by species
plt.figure(figsize=(10, 6))

species_colors = {'setosa': '#FF6B6B', 'versicolor': '#4ECDC4', 'virginica': '#45B7D1'}

for species in target_names:
    mask = df['Species'] == species
    plt.scatter(df[mask]['sepal length (cm)'], 
                df[mask]['petal length (cm)'],
                label=species,
                s=100,
                alpha=0.7,
                color=species_colors[species])

plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.title('Sepal Length vs Petal Length', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Correlation Analysis

Let's examine correlations between features. This helps us understand which features are related to each other.


```python
# Calculate correlation matrix
correlation_matrix = df.drop('Species', axis=1).corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("Correlation Matrix:")
print(correlation_matrix)
```

## Key Takeaways

1. **Dataset Shape**: 150 samples with 4 features and 3 classes
2. **Class Balance**: Each species has 50 samples (well-balanced)
3. **Feature Ranges**: Features have different scales, which may require normalization
4. **Feature Relationships**: Some features (like petal length and petal width) are highly correlated
5. **Class Separability**: Visual inspection suggests the classes can be separated using features
